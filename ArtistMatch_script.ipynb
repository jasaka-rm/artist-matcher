# ArtistMatch Project

The project includes:
- a small database of singers, authors, and albums
- a local database written directly in Python,
- an Excel database loaded with pandas,
- a short questionnaire,
- a basic matching system that recommends a singer to the user


---

```python
# PART 1 - DATA CLASSES

class Singer:
    """
    Represents one singer profile.
    This is the main class used by the recommendation system.
    """
    def __init__(self, name, genre, mood, era, energy, nationality, famous_song):
        self.name = name
        self.genre = genre
        self.mood = mood
        self.era = era
        self.energy = energy
        self.nationality = nationality
        self.famous_song = famous_song

    def show_info(self):
        print(f"Singer: {self.name}")
        print(f"Genre: {self.genre}")
        print(f"Mood: {self.mood}")
        print(f"Era: {self.era}")
        print(f"Energy: {self.energy}")
        print(f"Nationality: {self.nationality}")
        print(f"Famous song: {self.famous_song}")


class Author:
    """
    Represents one author in the database.
    Authors are included to make the project broader than a song-only recommender.
    """
    def __init__(self, name, genre, era, nationality, famous_book):
        self.name = name
        self.genre = genre
        self.era = era
        self.nationality = nationality
        self.famous_book = famous_book

    def show_info(self):
        print(f"Author: {self.name}")
        print(f"Genre: {self.genre}")
        print(f"Era: {self.era}")
        print(f"Nationality: {self.nationality}")
        print(f"Famous book: {self.famous_book}")


class Album:
    """
    Represents one album linked to a singer.
    """
    def __init__(self, title, singer_name, genre, year):
        self.title = title
        self.singer_name = singer_name
        self.genre = genre
        self.year = year

    def show_info(self):
        print(f"Album: {self.title}")
        print(f"Singer: {self.singer_name}")
        print(f"Genre: {self.genre}")
        print(f"Year: {self.year}")
```

```python
# PART 2 - DATABASE MANAGEMENT

import pandas as pd


class MediaLibrary:
    """
    Central class that stores all project data.
    It contains singers, authors, and albums.
    """
    def __init__(self):
        self.singers = []
        self.authors = []
        self.albums = []


    # OPTION A: local database directly in code -Local data** written directly in Python.

    def load_local_data(self):
        """Create a small built-in database for a simple classroom demo."""
        self.singers = [
            Singer("Taylor Swift", "pop", "emotional", "modern", "medium", "American", "Love Story"),
            Singer("Kendrick Lamar", "rap", "thoughtful", "modern", "high", "American", "HUMBLE."),
            Singer("Adele", "soul", "emotional", "modern", "low", "British", "Hello"),
            Singer("Ed Sheeran", "pop", "romantic", "modern", "medium", "British", "Shape of You"),
            Singer("Freddie Mercury", "rock", "energetic", "classic", "high", "British", "Bohemian Rhapsody"),
            Singer("Billie Eilish", "pop", "dark", "modern", "low", "American", "bad guy")
        ]

        self.authors = [
            Author("Jane Austen", "romance", "classic", "British", "Pride and Prejudice"),
            Author("George Orwell", "dystopia", "classic", "British", "1984"),
            Author("Victor Hugo", "historical", "classic", "French", "Les Miserables"),
            Author("J.K. Rowling", "fantasy", "modern", "British", "Harry Potter")
        ]

        self.albums = [
            Album("1989", "Taylor Swift", "pop", 2014),
            Album("30", "Adele", "soul", 2021),
            Album("DAMN.", "Kendrick Lamar", "rap", 2017),
            Album("Divide", "Ed Sheeran", "pop", 2017),
            Album("A Night at the Opera", "Freddie Mercury", "rock", 1975),
            Album("When We All Fall Asleep, Where Do We Go?", "Billie Eilish", "pop", 2019)
        ]


    # OPTION B: load the database from an Excel file with pandas

    def load_from_excel(self, excel_file_path):

      df = pd.read_excel(excel_file_path)

      df.columns = df.columns.str.strip()

      self.singers = []
      self.authors = []
      self.albums = []

      for _, row in df.iterrows():
          self.singers.append(
              Singer(
                  row["name"],
                  row["genre"],
                  row["mood"],
                  row["era"],
                  row["energy"],
                  row["nationality"],
                  row["famous_song"]
              )
          )


    # Simple search tools
    def find_singers_by_genre(self, genre):
        return [singer for singer in self.singers if singer.genre.lower() == genre.lower()]

    def find_albums_by_singer(self, singer_name):
        return [album for album in self.albums if album.singer_name.lower() == singer_name.lower()]

    def show_database_summary(self):
        print("Database summary")
        print("----------------")
        print(f"Number of singers: {len(self.singers)}")
        print(f"Number of authors: {len(self.authors)}")
        print(f"Number of albums: {len(self.albums)}")
```

```python
# PART 3 - QUESTIONNAIRE

class UserQuestionnaire:
    """
    Collects the user's preferences.
    """
    def ask_questions(self):
        answers = {}
        answers["genre"] = input("What music genre do you prefer? (pop / rap / soul / rock)").strip().lower()
        answers["mood"] = input("What mood do you prefer? (emotional / romantic / energetic / dark / thoughtful) ").strip().lower()
        answers["era"] = input("Do you prefer modern or classic artists? ").strip().lower()
        answers["energy"] = input("Do you prefer low, medium, or high energy music? ").strip().lower()
        return answers
```

```python
# PART 4 - MATCHING ALGORITHM

class SingerMatcher:
    """
    Compares user answers with singer profiles.
    """
    def __init__(self, singers):
        self.singers = singers

    def score_singer(self, singer, answers):
        score = 0

        if singer.genre.lower() == answers.get("genre", ""):
            score += 1
        if singer.mood.lower() == answers.get("mood", ""):
            score += 1
        if singer.era.lower() == answers.get("era", ""):
            score += 1
        if singer.energy.lower() == answers.get("energy", ""):
            score += 1

        return score

    def find_best_match(self, answers):
        best_singer = None
        best_score = -1

        for singer in self.singers:
            current_score = self.score_singer(singer, answers)
            if current_score > best_score:
                best_score = current_score
                best_singer = singer

        return best_singer, best_score

    def show_top_matches(self, answers):
        scored_results = []

        for singer in self.singers:
            current_score = self.score_singer(singer, answers)
            scored_results.append((singer, current_score))

        scored_results.sort(key=lambda item: item[1], reverse=True)

        print("Top singer matches:")
        for singer, score in scored_results[:3]:
            print(f"- {singer.name}: {score}/4")
```

```python
# PART 5 - DEMO WITH LOCAL DATA

library = MediaLibrary()
library.load_local_data()
library.show_database_summary()

print("----------------")
print("Pop singers in the database:")
for singer in library.find_singers_by_genre("pop"):
    print("-", singer.name)

print("Albums by Ed Sheeran:")
for album in library.find_albums_by_singer("Ed Sheeran"):
    print("-", album.title)
```

```python
# QUICK MATCH DEMO

# Run questionnaire
questionnaire = UserQuestionnaire()
user_answers = questionnaire.ask_questions()

matcher = SingerMatcher(library.singers)
best_singer, best_score = matcher.find_best_match(user_answers)


print(f"Best match: {best_singer.name} with a score of {best_score}/4")
print(f"Famous song: {best_singer.famous_song}")
```
